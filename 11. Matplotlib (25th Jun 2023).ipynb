## Multi Line graph

```python
y1=[1,2,30,4,5]
y2=[10,20,30,40]
plt.plot(y1,color="r",marker="*")
plt.plot(y2,color="b")
plt.show()
```

```python
y1=[1,2,30,4,5]
y2=[10,20,30,40]
plt.plot(y1,color="r",marker="*")
plt.plot(y2,color="b")
plt.xlabel("X Axis",color="r",fontsize=20)
plt.ylabel("Y Axis",color="b",fontsize=20)
plt.title("Multi Line Ghraph",color="green")
plt.show()
```

```python
y1=[1,2,30,4,5]
y2=[10,20,30,40]
plt.plot(y1,color="r",marker="*",lw=3,ls="--")
plt.plot(y2,color="b")
plt.xlabel("X Axis",color="r",fontsize=20)
plt.ylabel("Y Axis",color="b",fontsize=20)
plt.title("Multi Line Ghraph",color="green")
plt.show()
```

## Subplot

```python

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2,3, 2)
plt.plot(x,y)

#plot 3:
x = np.array([0, 1, 2, 3])
y = np.array([30, 20, 30, 40])

plt.subplot(2,3, 3)
plt.plot(x,y)

plt.show()
```

##### Adding subtitle

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])


plt.subplot(2, 2, 1)
plt.plot(x,y)

plt.grid(color="Green")
plt.title("Sub Ploting 1")
#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2,2, 2)
plt.plot(x,y)

plt.grid(color="lightblue")
plt.title("Sub Ploting 2")
#plot 3:
x = np.array([0, 1, 2, 3])
y = np.array([30, 20, 30, 40])

plt.subplot(2,2, 3)
plt.plot(x,y)

plt.grid(color="k", axis="x")
plt.title("Sub Ploting 3")
#plot 4:
x = np.array([0, 1, 2, 3])
y = np.array([30, 20, 30, 40])

plt.subplot(2,2, 4)
plt.plot(x,y)

plt.grid(color="hotpink", axis="y")
plt.title("Sub Ploting 3")

plt.suptitle("Multi Plots", fontsize=20,color="r")

plt.show()
```
